## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X=sc.fit_transform(df)
X
```

```python
X.shape
```

```python
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy
```

**constructing dendogram With single & Complete Linkage**

```python
# Single linkage
plt.figure(figsize=(10,7))
plt.title('Crime data dendogram with Single Linkage')

lk = hierarchy.linkage(X, method='single')
dendrogram = hierarchy.dendrogram(lk)
```

```python
# Complete linkage
plt.figure(figsize=(10,7))
plt.title('Crime data dendogram with Complete Linkage')

lk = hierarchy.linkage(X, method='complete')
dendrogram = hierarchy.dendrogram(lk)
```

```python
# Average linkage
plt.figure(figsize=(10,7))
plt.title('Crime data dendogram with Average Linkage')

lk = hierarchy.linkage(X, method='average')
dendrogram = hierarchy.dendrogram(lk)
```

## Model Building

```python
#create cluster
from sklearn.cluster import AgglomerativeClustering
```

```python
# Here we'll use complete linkage and 4 clusters
hc = AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='complete')
ypred = hc.fit_predict(X)
```

```python
df['clusters'] = ypred
df
```

```python
df.groupby('clusters').mean()
```

# k-means

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=5)
```

```python
yp=km.fit_predict(X)
yp
```

```python
#cluster centre
km.cluster_centers_
```

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(X)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.axvline(x = 4, linestyle = '--', color = 'red')
plt.xlabel('Number of cluster')
plt.ylabel('inertial values')
plt.show()
```

```python
model = KMeans(n_clusters=4)
ypred = model.fit_predict(X)
ypred
```

```python
#Silhouette score
from sklearn.metrics import silhouette_score
silhouette_score(X,ypred)
```

```python
Clust=pd.DataFrame(model.labels_,columns=['wcss'])
y3=pd.concat([df,Clust], axis=1)
y3
```

```python
y3.iloc[:,1:5].groupby(y3.wcss).mean()
```

# DBSCAn clustering

```python

#standerization
x=df.iloc[:,1:]
ss= StandardScaler()
x_ss = ss.fit_transform(x)
x_ss

```

```python
from sklearn.cluster import DBSCAN
DBSCAN()
```

```python
dbscan=DBSCAN(eps=1,min_samples=6)
dbscan.fit(x_ss)
```

```python
dbscan.labels_          #(-1 are noise)
```

```python
c1=pd.DataFrame(dbscan.labels_,columns=['DBclust'])
c1
```

```python
DBclust=pd.concat([df,c1],axis=1)
DBclust
```

```python
noisedata= DBclust[DBclust['DBclust']==-1]
print(noisedata)
```

```python
finaldata= DBclust[DBclust['DBclust']==0]
print(finaldata)
```

```python
DBclust.iloc[:,1:5].groupby(DBclust.DBclust).mean()
```

```python
from sklearn.preprocessing import MinMaxScaler
MM = MinMaxScaler()
MM_x=MM.fit_transform(x)
```

```python
MM_x.shape
```

```python
#construct dendogram
plt.figure(figsize=(10,7))
plt.title('Airline data dendogram')
dend = sch.dendrogram(sch.linkage(MM_x,method='average'))
```

```python

#create cluster
ac = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='single')
```

```python
y5
```

```python
#k-means clustering
```

```python
from sklearn.cluster import KMeans
KMeans()
```

```python
kmeans=KMeans(n_clusters=5)
kmeans=kmeans.fit(MM_x)
```

```python
kmeans.inertia_
clust=[]
for i in range(1,11):
   kmeans = KMeans(n_clusters=i,random_state=0)
   kmeans.fit(MM_x)
   clust.append(kmeans.inertia_)
   clust
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),clust)
plt.title('Elbow Method')
plt.xlabel('Number of cluster')
plt.ylabel('inertial values')
plt.show()
```

```python
model= KMeans(n_clusters=5)
model.fit(MM_x)
model.labels_
```

```python
Clust=pd.DataFrame(model.labels_,columns=['kClust'])
y6=pd.concat([fd,Clust], axis=1)
y6
```

```python
y6.iloc[:,1:12].groupby(y6.kClust).mean()
```
